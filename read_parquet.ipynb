```python
import geopandas as gpd
import sys
import pandas as pd
```

```python
# Specify the path to your Parquet file
file_path = 'Data/gdf_23-11-01_23-12-01.parquet'
try:
    # Load the Parquet file using GeoPandas
    gdf = gpd.read_parquet(file_path)
    print("File loaded successfully!")

    # Print the total number of rows
    print("Total number of rows:", len(gdf))
    
    # Print the column names
    print("Column names:", gdf.columns.tolist())

except Exception as e:
    print("An error occurred:", e)

# Check the current CRS
print("Current CRS:", gdf.crs)
print(gdf)
```

```python
# Convert the parquet file to text
all_data = gdf.to_string()

# Open a text file to redirect output
with open('new_cleaned/cleaned_november.txt', 'w') as file:
    # Redirect standard output to the file
    sys.stdout = file

    # The output of this print statement will go directly to 'output.txt'
    print(all_data)

    # Reset standard output to default
    sys.stdout = sys.__stdout__
```

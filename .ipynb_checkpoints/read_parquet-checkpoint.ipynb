```python
import geopandas as gpd
import sys
import pandas as pd
```

```python
# Specify the path to your Parquet file
file_path = 'Data/gdf_23-11-01_23-12-01.parquet'
try:
    # Load the Parquet file using GeoPandas
    gdf = gpd.read_parquet(file_path)
    print("File loaded successfully!")

    # Print the total number of rows
    print("Total number of rows:", len(gdf))
    
    # Print the column names
    print("Column names:", gdf.columns.tolist())

except Exception as e:
    print("An error occurred:", e)
```

```python
# Convert the first 50 rows to a string
first_5000_rows = gdf.head(5000).to_string()
#all_data = gdf.to_string()

# Open a text file to redirect output
with open('first_5000_rows.txt', 'w') as file:
    # Redirect standard output to the file
    sys.stdout = file

    # The output of this print statement will go directly to 'output.txt'
    print(first_5000_rows)

    # Reset standard output to default
    sys.stdout = sys.__stdout__
```

```python


# Assuming the file is a CSV formatted text file
file_path = 'Data/first_5000_rows.txt'

# Load the file with pandas to inspect the columns
df = pd.read_csv(file_path, nrows=0)  # nrows=0 loads no data, just columns

# Print the column names
print("Column names:", df.columns.tolist())
```

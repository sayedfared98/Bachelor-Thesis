```python
import geopandas as gpd
import pandas as pd
import plotly.express as px
```

```python
# Load the GeoDataFrame
gdf = gpd.read_parquet('Data/gdf_23-11-01_23-12-01.parquet')
```

Statistical Summaries

```python
#For numerical data like speed, longitude, latitude, and linkPos
print(gdf.describe())
```

```python
#Count Unique Values for Categorical Data
#For categorical data like route, vehicle, and trip
#print(gdf['route'].value_counts())
#print(gdf['vehicle'].value_counts())
print(gdf['trip'].value_counts())
```

Data Visualization

```python
# Ensure utcTime is a datetime type
gdf['utcTime'] = pd.to_datetime(gdf['utcTime'])

# Create a line plot
fig = px.line(gdf, x='utcTime', y='speed', title='Speed Over Time', labels={'speed': 'Speed (km/h)'})
fig.update_xaxes(title_text='Time')
fig.update_yaxes(title_text='Speed (km/h)')
fig.show()
```
